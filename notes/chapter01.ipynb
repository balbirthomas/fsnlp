## Notes

1. Three central problems in linguistic science
+ How do humans acquire, produce, use and understand language ?
+ How do linguistic utterances represent the world and relate to it ?
+ What are the linguistic structures by which language communicates ?

2. The common conception is that there are rules (Grammar) which structure linguistic expression for the purpose of communication. However these "Grammars" fail to characterize the full range of linguistic expression actually observed. In the words of Edward Sapir - "All grammars leak".

3. Studying statistical patterns in linguistic expression provides an empirical foundation for the study of language. Statistical analysis use a body of text, the corpus, as a representation of real language usage. This is often referred to as "Corpus Linguistics".

4. It is likely that innate biological features of the human brain predispose us towards the acquisition of language and its use. This may be used argue that there exists an internal representation of language (I language) distinct from its external manifestation (E language) used in communication.

5. Generative (Chomskyan) linguists seek to understand the cognitive basis of language by formulating and testing models of its internal subconscious representation, the I language.

6. Generative linguists distinguish between Linguistic Competence and Linguistic Performance. *Linguistic competence* is the innate and unconscious knowledge underlying the language. *Linguistic Performance* is the actual use of the language in concrete situations. 

6. Other schools of linguistics such as [usage based models of language](https://en.wikipedia.org/wiki/Usage-based_models_of_language) reject the notion that one can isolate linguistic competence and study it in isolation. Some linguists (Empiricists) are focused on describing the actual use of language.

7. Engineering solutions to practical problems favour Statistical Natural Language Processing approaches because they are better at automatic learning (Knowledge Induction), better at disambiguation and provide an empirical foundation for the science of linguistics.

8. Chomskyan linguists are concerned with if sentences in a corpus do or do not conform to categorical principle of sentence construction, for example does a sentence conform to the grammar i.e. is structurally well formed. Statistical NLP practitioners would like to ascribe probabilities to linguistic events and categorize the sentences as "likely" or "unusual".

9. Patterns found through statistical analysis of linguistic corpora reveal the syntactic structure of language. Patterns of word use also provide insight into their semantics.

10. Generative linguists abstract away from concrete expressions of language and try to describe a *competence grammar* that underlies the language, differentiating between grammatical and ungrammatical linguistic expressions. This *syntactic gramaticality* is a binary choice.

11. It is possible for linguistic utterances to be grammatically correct but are still "unusual" and may "seem odd" for example "Colorless green ideas sleep furiously". This *conventionality* in modes of linguistic utterances is not explained by the binary classification of sentences into grammatical and ungrammatical.

12. It is possible to construct sentences that human beings find difficult to classify as grammatical or ungrammatical. For example
    - "John I believe Sally said Bill believed Sue saw." is grammatical.
    - "What did Sally whisper that she had secretly read?" is ungrammatical.
    - "Who did Jill think said John saw him?" is grammatical.
    - "The boys read Mary’s stories about each other." is ungrammatical.

13. Several observations suggest that the *categorical view of language* may be sufficient for many purposes but is an approximation with limitations that struggles to explain non-categorical phenomenon, for example
    - Languages evolve both in their syntax and semantics. Statistically such evolution may be explained by gradual variations over time but it is harder to reconcile with the binary dichotomy of grammatical vs ungrammatical expressions.
    - Languages show sociolinguistic variations.
    - Parts of speech are expected to be categorically distinct. Many English words may serve multiple parts of speech but in different contexts. Some words however blend these categories (parts of speech) for example the word "nearer" simultaneously shows properties of both adjectives and prepositions in the sentence "We live nearer to the lake than you thought".
    - The part of speech a word serves may change. For example in the sentence "What sort of animal made these tracks?" the word "sort" serves as noun with "of" as a following preposition introducing a prepositional phrase. But there is another usage of the phrase "sort of" as in "He sort of understood what was going on" where it serves as a degree modifier. Historically "sort" was a noun and there is no evidence of change in its usage being externally influenced. Language changes may happen suddenly for example under external influence but typically they are gradual. The details of gradual change can only be made sense of by examining frequencies of use and being sensitive to varying strengths of relationships, and this type of modeling requires statistical, as opposed to categorical, observations.

14. The world is uncertain and provides incomplete information. In such a context cognition and decision making require a quantitative framework to process uncertain and incomplete information for instance by considering probabilities of events. It can further be argued that since human cognition is probabilistic, language must also therefore be probabilistic, being an integral part of cognition. Besides the cognitive process used for language are similar to those for processing other kinds of sensory input and knowledge.

15. Noam Chomsky has argued that probability theory is inappropriate for formalizing the notion of grammaticality. Since computing the probability of sentences from a corpus of utterances would assign the same low probability to all unattested sentences, grammatical and ungrammatical ones alike, he argues, it does not hence account for linguistic productivity.

16. Statistical NLP provides methods for deriving the probability estimates for unseen linguistic events. Hence challenging the premise that all unattested sentences, grammatical and ungrammatical ones alike will be assigned the same low probability.

17. In addition to insights into the syntactic structure of language, Statistical NLP methods can also address issues of semantics i.e. "meaning". Statistical NLP treats "meaning" as residing in the distribution of contexts over which words and utterances are used. This is known as the *use theory of meaning.

18. For a given grammar a sentence may have alternative syntactic analysis know as *parses*. Each syntactic analysis is one way of ascribing parts of speech to the components (words and phrases) of the sentence. As sentences get longer and grammars get more complex the number of possible parses for a given sentence increases. Hence any practical natural language processing (NLP) system needs to be good at disambiguation decisions of word sense, word category, syntactic structure, and semantic scope.

19. The goal of maximizing coverage while minimizing resultant ambiguity is fundamentally inconsistent with symbolic (as opposed to statistical) NLP systems, where extending the coverage of the grammar to obscure constructions simply increases the number of undesired parses for common sentences and vice versa. For example use of *selection restriction* in symbolic NLP for parsing and disambiguation can easily fail. As an example suppose we place the restriction that the verb like swallow requires an animate being as its subject and a physical object as its object. Such a restriction would disallow common metaphorical usage of the word "swallow" as in "I swallowed his story, hook, line, and sinker". Statistical NLP seeks to solve such problems by automatically learning lexical and structural preferences from corpora. Further probabilistic information of collocation of words (*collocation knowledge*) may be used to gain insight into semantic relationships and resolve ambiguities.

20. Statistical models tend to be robust, generalize well and behave gracefully in the presence of errors and new data.

21. A common first task in characterizing a corpus is to compute its *word frequencies* (count of occurrences). In doing so one distinguishes between *tokens* which are individual occurrences of something and *types*, the different things present. Word frequencies are useful in Cryptography and Cryptanalysis. Word frequency distributions vary across corpora. The ratio of tokens to types is the average frequency with which each type is used.

22. One of the challenges of Statistical NLP is that it is harder to predict much about he behavior of words that are rarely observed in the corpus. This problem can persist even with large corpora.

23. If we count up how often each word (type) of a language occurs in a large corpus, and then list the words in order of their descending frequency of occurrence, we can explore the relationship between the frequency of a word $f$ and its position in the list, known as its rank $r$ . **Zipf’s law** says that

$$ f \propto \frac{1}{r} $$

This implies there exists a constant $k$ such that $f \cdot r = k$.

24. Benoit Mandelbrot improved upon Zif's law by providing the following formula (also hyperbolic) that is a closer fit to empirical distribution of words.

$$ f = P (r + ρ)^{-B} $$

which in logarithmic form is

$$ \log f = \log P - B \log(r + \rho) $$

where $P$, $B$ and $\rho$ are parameters of a text, that collectively measure the richness of the text's use of words. A double logarithmic plot of $f$ vs $r$ using this formula approximates a straight line with slope $-B$ for large $r$. If $B=1$ and $\rho=0$ then Mandelbrot's formula simplifies to Zipf's law. Zipf's law captures the insight that frequency based approaches to the study of language is hard because almost all the words are rare.

25. Zipf also proposed that the number of meanings $m$ of a word is proportional to it frequency $f$ and the relation is given by

$$ m \propto \sqrt{f} $$

which can be transformed using Zipf's law to

$$ m \propto \frac{1}{\sqrt{r}} $$

26. There is a tendency for words in documents to clump i.e. occurrences of the word tend to be found in proximity with each other. Let the interval size $I$ be the number of lines or pages between two occurrences of a word. Let $F$ be the frequency of the interval size $I$ for a specific word then Zipf observed that

$$ F \propto I^{-p} $$

where $p$ varied between 1 and 1.3.

27. Zipf’s also observed that there is an inverse relationship between the frequency of words and their length, that the greater the frequency of a word or morpheme, the greater the number of different permutations (roughly, compounds and morphologically complex forms) it will be used in. Further observations by Zipf noted empirical relations in historical change and the frequency of phonemes.

28. Zipf's law is a specific example of a *Power Laws* ($y = k \cdot x^n$ for some constants $k$ and $n$) that are commonly observed in descriptions of many natural phenomenon. There are many statistical distributions that obey power laws when Zipf's procedure (counting events and ranking them by descending frequency) is applied to them.

29. Randomly generated text exhibit Zipf's law or rather its generalization by Mandelbrot since
    + The probability of generating a $n$ character word (using the English alphabet of 26 characters and 1 blank (space) is $\left(\frac{26}{27}\right)^n \frac{1}{27}$.
    + There are 26 times more words of length $n+1$ than there are of length $n$
    + There is a constant ratio by which words of length $n$ are more frequent than words of length $n+1$.

30. A *Collocation* is any turn of phrase or accepted usage where somehow the whole is perceived to have an existence beyond the sum of the parts. For example "disk drive" or "make up". Collocations typically have specialized meaning or are idiomatic but not necessarily so. In fact any expression that people repeat because they have heard others use it is a candidate for collocation. Collocations have applications in machine translation and information retrieval. In machine translation words may be translated differently according to the collocations they occur in. Information retrieval system may want to only index "interesting" phrases such as collocations. Lexicographers may also be interested in collocations because of their specialized meaning and independent existence as a multi-word unit. Collocations may be discontinuous, for example "make \[something\] up".

31. Identifying collocations in copora can be challenging since the naive approach searching for frequently used 2 (*bigram*) or more (*n-gram*) word yields poor results. This is because it does not take into consideration and normalize for frequency of the individual words themselves. Further improvement in collocation detection may be effected by removing those collocations that have parts of speech (syntactic categories) that are rarely associated with interesting collocations.

32. *Key Word in Context* is a method of displaying the syntactic frame in which a word (keyword) appears. This is done by lining up the occurrence of the word one above another in a corpus of sentences. The context of the word is then visible on its left and right. The understanding gained of syntactic patterns using Key Words in Context is useful for building statistical parsers.

```python
import collections
import string

import numpy as np
import scipy as sp

from IPython.display import HTML, display
from matplotlib import pyplot as plt
```

```python
def display_html_table(data, headers):
    """
    Display a table in HTML format.

    Arguments:
        data: a list of tuples containing the table data
        headers: a list of column headers
    """
    html = '<table border="1"><tr>'
    for header in headers:
        html += f'<th>{header}</th>'
    html += '</tr>'
    for row in data:
        html += '<tr>'
        for cell in row:
            html += f'<td>{cell}</td>'
        html += '</tr>'
    html += '</table>'
    display(HTML(html))

def count_words(text):
    """
    Count words in a text string.

    Arguments:
        text: a string containing the text to be analyzed

    Returns:
        A dict keyed by word, with word counts
    """
    chunks = text.split()
    punctuation_removed = [word.strip(string.punctuation) for word in chunks]
    word_list = [word.lower() for word in punctuation_removed if word]
    word_counts = collections.Counter(word_list)
    return word_counts

def zipf(r, k=1):
    """Zipf's law function.

    Arguments:
        r: the rank of the word
        k: a constant (default is 1)

    Returns:
        The frequency of the word according to Zipf's law
    """
    return k / r

def mandelbrot(r, P=1, B=1, rho=1):
    """Mandelbrot's generalization of Zipf's law.

    Arguments:
        r: the rank of the word
        P: a constant (default is 1)
        B: a constant (default is 1)
        rho: a constant (default is 1)

    Returns:
        The frequency of the word according to Mandelbrot's law
    """
    return P / ((r + rho) ** B)
```

```python
with open('../data/tom_sawyer_1972.txt', 'r') as file:
    text = file.read()

MAX_ITEMS = 15

word_frequency = count_words(text)
display_html_table(word_frequency.most_common(MAX_ITEMS), ["Word", "Frequency"])
```

```python
ranked_frequencies = sorted(word_frequency.values(), key=lambda x: x, reverse=True)
ranks = range(1, len(ranked_frequencies) + 1)
zipf_parameters = sp.optimize.curve_fit(zipf, ranks, ranked_frequencies)[0]
plt.figure(figsize=(10, 6))
plt.loglog(ranks, ranked_frequencies, 'o')
plt.loglog(ranks, zipf(ranks, *zipf_parameters), label="Zipf's Law", color='red')
plt.title("Zipf's Law: Word Frequency vs. Rank")
plt.xlabel("Rank")
plt.ylabel("Frequency")
plt.grid(True)
plt.legend()
plt.show()
```

$$\begin{align*}

```python

# Here is the longest dictionary word according to the Oxford English Dictionary.
LONGEST_DICTIONARY_WORD = "pneumonoultramicroscopicsilicovolcanoconiosis"

def random_word(length):
    """Generate a random word of a given length.

    Arguments:
        length: the length of the word to be generated

    Returns:
        A random word of the specified length
    """
    letters = string.ascii_lowercase
    return ''.join(np.random.choice(list(letters), size=length))

def generate_random_words(num_words, max_length=0):
    """Generate a list of random words.

    The length of each word is determined by a geometric distribution,
    which models the number of non-space characters until the first space.
    The parameter p=1/27 for the geometric distribution is chosen to 
    reflect the average word length in English (26 letters + 1 space).
    The max_length parameter can be used to limit the maximum length of
    the generated words.

    Arguments:
        num_words: the number of random words to generate
        max_length: the maximum length of the generated words (default is 0, which means no limit)

    Returns:
        A list of random words
    """
    words = []
    for _ in range(num_words):
        # number of non-space characters till the first space
        # determines the word length
        word_length = np.random.geometric(p = 1/27)
        if max_length > 0 and word_length > max_length:
            word_length = max_length
        words.append(random_word(word_length))
    return words

def blank_limited_random_words(num_words):
    """Generate a list of random words with a maximum length.

    Arguments:
        num_words: the maximum number of random words to generate

    Returns:
        A list of at most `num_words` random words
    """
    words = "".join(np.random.choice(list(string.ascii_lowercase + ' '), size=num_words * len(LONGEST_DICTIONARY_WORD))).split()
    limit = min(len(words), num_words)
    return words[:limit]
```

```python
random_words = generate_random_words(71000, max_length=0) # there are about 71,000 words in the Tom Sawyer text
random_word_counts = collections.Counter(random_words)
random_word_ranked_frequencies = sorted(random_word_counts.values(), key=lambda x: x, reverse=True)
random_word_ranks = range(1, len(random_word_ranked_frequencies) + 1)
random_word_zipf_parameters = sp.optimize.curve_fit(zipf, random_word_ranks, random_word_ranked_frequencies)[0]

plt.figure(figsize=(10, 6))
plt.loglog(random_word_ranks, random_word_ranked_frequencies, 'o')
plt.loglog(random_word_ranks, zipf(random_word_ranks, *random_word_zipf_parameters), label="Zipf's Law", color='red')
plt.title("Random Words: Word Frequency vs. Rank")
plt.xlabel("Rank")
plt.ylabel("Frequency")
plt.grid(True)
plt.legend()
plt.show()
```

### Exercise 1.5

Think about ways of identifying collocations that might be better than the methods used in this chapter.

A naive (simplistic) method to detect collocations is implemented below. This method does nothing more than count the frequency of occurrence of tandem word groups.

```python
def sanitize_word(word):
    """Sanitize a word by removing punctuation and converting to lowercase.

    Arguments:
        word: the word to be sanitized
    """
    return word.translate(str.maketrans('', '', string.punctuation)).strip().lower()

def collocations(text, window_size=2):
    """Extract collocations from a text.

    Arguments:
        text: a string containing the text to be analyzed
        window_size: the number of words to consider on either side of a target word (default is 2) 
    Returns:
        A dict keyed by collocation, with counts
    """
    collocations = collections.Counter()
    words = text.split()
    for i in range(len(words)):
        target_word = sanitize_word(words[i])
        context_start = max(0, i - window_size - 1) # -1 to excludes the target word from the left context
        context_end = min(len(words), i + window_size) # target word is included in the right context
        left_context = [sanitize_word(word) for word in words[context_start:i]]
        right_context = [sanitize_word(word) for word in words[i+1:context_end]]
        left_collocation = ' '.join(left_context + [target_word])
        right_collocation = ' '.join([target_word] + right_context)
        collocations[left_collocation] += 1
        collocations[right_collocation] += 1
    return collocations
```

As shown by the results below the simplistic collocation detection algorithm performs poorly because the most common tandem word repeats are just common language usage patters and not collocations.

```python
with open('../data/tom_sawyer_1972.txt', 'r') as file:
    text = file.read()
    collocations_count = collocations(text)
display_html_table(collocations_count.most_common(10), ["Collocation", "Frequency"])
```

One way we may try to improve the collocation detection procedure is by eliminating commonly occurring word patterns that are not collocations by examining the word frequency of the words in the corpus and eliminating candidate collocations that have any common words.

```python
def filter_collocations(collocations, common_words):
    """Filter collocations to remove those with common words.

    Arguments:
        collocations: a dict keyed by collocation, with counts
        common_words: a set of common words to be filtered out

    Returns:
        A dict of filtered collocations that meet the uncommon word criteria
    """
    filtered_collocations = {}
    for collocation, count in collocations.items():
        words = collocation.split()
        if all (word not in common_words for word in words):
            filtered_collocations[collocation] = count
    return filtered_collocations

def common_words(text, num_common=100):
    """Get the most common words from a text.

    Arguments:
        text: a string containing the text to be analyzed
        num_common: the number of common words to return (default is 100)
    """
    words = text.split()
    word_counts = collections.Counter(words)
    return set(word for word, count in word_counts.most_common(num_common))

with open('../data/tom_sawyer_1972.txt', 'r') as file:
    text = file.read()
    collocations_count = collocations(text)
    common_words_set = common_words(text)
    filtered_collocations = filter_collocations(collocations_count, common_words_set)
display_html_table(collections.Counter(filtered_collocations).most_common(10), ["Collocation", "Frequency"])
```

### Exercise 1.7

Write a program to produce KWIC displays from a text file. Have the user be able to select the word of interest and the size of the surrounding context.

```python
def kwic(sentences, word, max_context=5):
    """Generate a KWIC (Key Word In Context) index for a given word in a list of sentences.

    Arguments:
        sentences: a list of sentences to be analyzed
        word: the word for which to generate the KWIC index
        max_context: the maximum number of words to include on either side of the target word (default is 5)

    Returns:
        A list of tuples containing the context of the specified word in each sentence
    """
    concordances = []
    for sentence in sentences:
        words = sentence.split()
        if word in words:
            index = words.index(word)
            left_context = ' '.join(words[max(0, index - max_context):index])
            right_context = ' '.join(words[index + 1:index + 1 + max_context])
            concordances.append((left_context, word, right_context))
    return concordances
```

```python
with open('../data/tom_sawyer_1972.txt', 'r') as file:
    text = file.read()
sentences = text.split('.') # poor man's sentence splitter
concordances = kwic(sentences, 'showed')
max_concordances = 10 # limit the number of concordances to display
display_html_table(concordances[:max_concordances], ["Left Context", "Word", "Right Context"])
```
