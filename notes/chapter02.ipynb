## Notes

1. *Probabiliy Theory* is concerned with predicting how likely it is that an event will occur.

2. An *experiment* or a *trial* is the process by which an observation is made.

3. A specific outcome (observation) of an experiment is known as *basic outcome* or *sample point*.

4. The collection of all possible outcomes for an experiment is as set called the *sample space* $\Omega$.

5. An *event* is a subset of the sample space $\Omega$.

6. Both the entire sample space $\Omega$ and the empty set $\emptyset$ are also *events*. The empty set represents an impossible event. For the purposes of these notes unless explicitly stated otherwise the sample space will always be denoted by $\Omega$.

7. Given any set $X$ the set of all subsets of $X$, denoted as $2^X$, is known as the *power set of $X$*.

8. Let $X$ be any set. A collection of subsets $\mathcal{F} \subset 2^X$ is called *$\sigma$-algebra* or *$\sigma$-field* if it satisfies the following conditions
- $X \in \mathcal{F}$
- If $Y \in \mathcal{F}$ then $Y \backslash \mathcal{F} \in \mathcal{F}$ where $Y \backslash \mathcal{F}$ is the complement of $Y$ relative to $\mathcal{F}$.
- Let $Y_i \in \mathcal{F}$, $i = 1, 2, 3 \ldots$ be a countable collection, then their union is also in $\mathcal{F}$ ( $\cup U_i \in \mathcal{F}$ ).

9. By implication the $\sigma$-field is also closed under countable intersection and the power set itself is a $\sigma$-field.

10. Given a $\sigma$-field $\mathcal{F}$, a *Probability Function* or *Probability Distribution* $P$ is a function $P : \mathcal{F} \to [0, 1]$ such that

- $P(\Omega) = 1$
- For *disjoint sets* $A_i \in \mathcal{F}$ (i.e $A_j \cap A_k = \emptyset$ for $j \neq k$)

$$P \left( \cup_{i=1}^\infty A_i \right) = \Sigma_{i=1}^\infty P(A_i)$$

Note $[0,1]$ is the closed interval between $0$ and $1$ inclusive.

11. A *Probability Space* ($\Omega, \mathcal{F}, P$) consists of a sample space $\Omega$, a $\sigma$-field $\mathcal{F}$ and a probability function $P$.

12. A sample space $\Omega$ with events that are equally likely is said to have a *uniform distribution*. For such a space the probability distribution $P$ is given by
$$P(A) = \frac{|A|}{|\Omega|}$$
where $A \subset \Omega$ is an event and the number of elements in any set $X$, known as the *cardinality* of $X$ is denoted as $|X|$.

13. Computing probabilities of events in discrete sample spaces often require the use of *Counting Techniques* described below

- *Fundamental Rule of Counting* : Suppose there are $n$ events each ($i=1 \ldots n$) having $k_i$ possible outcomes. Then the total number of possible outcomes for the collection of all the events is $k_1 \cdot k_2 \ldots \cdot k_n$.

- *Permutations* : Suppose there are $n$ distinguishable things out of which we wish to select $k$ things and arrange them as an ordered sequence. Then the number of different arrangements that can be made is known as the *permutations* of $n$ things taken $k$ at a time and is denoted by ${}_nP_k$

$$ {}_nP_k = \frac{n!}{(n-k)!} $$

where $n!$ is the factorial function ($!:\mathbb{N} \to \mathbb{N}$) defined recursively by $n! = n \cdot (n-1)!$  and $0! = 1$.

- Suppose there are $n$ things of $k$ distinguishable types ($k \leq n$) and we seek to select $n_i$ ($i=1 \ldots k$) things of the $i^{th}$ kind and arrange them as an ordered sequence then the number of different arrangements that can be made is

$$ \frac{n!}{n_1! n_2! \cdots n_k!} $$

- *Combinations* : Suppose there are $n$ distinguishable things and we seek to select a subset of $k$ things. The number of ways in which this selection can be made is known as the *combinations* ${}_nC_k$ of $n$ things taken $k$ at a time and is given by

$$ {}_nC_k = \frac{n!}{k! (n-k)!} $$

Combinations differ from permutations in that the ordering of $k$ selected things is disregarded in combinations. There are $k!$ such orderings. Trivially ${}_nP_k = k! \cdot {}_nC_k$.

14. The *Conditional Probability* of an event $A$ updates its probability in the light of new information provided by the occurrence of another event $B$. The probability of the event before taking into consideration the additional information is known as the *Prior probability* and that after taking into consideration the additional information is known as the *Posterior probability*.

15. Suppose $A, B \subset \Omega$ are any two events in a discrete sample space $\Omega$

$$ P(A \cap B) = \frac{|A \cap B|}{|\Omega|} $$

and

$$ P(B) = \frac{|B|}{|\Omega|} $$

This implies

$$ \frac{P(A \cap B)}{P(B)} = \frac{|A \cap B|}{|B|} $$

If the sample space was restricted to $B$ then ratio of $|A \cap B|$ and $|B|$ is the probability of $A$ in this restricted sample space. This motivates the following definition.

The *Conditional Probability* $P(A | B)$ of an event $A$ given that event $B$ (assuming $P(B) > 0$) has occurred is

$$P(A | B) = \frac{P(A \cap B)}{P(B)}$$

Even if $P(B)=0$

$$P(A \cap B) = P(B) P(A | B) = P(A) P(B | A)$$

In this form the relation is known as the *Multiplication rule*. There are two other basic rules

*Addition Rule*

$$ P(A \cup B) = P(A) + P(B) - P(A \cap B) $$

*Complement Rule*

$$ P(A^c) = 1 - P(A) $$

where $A^c = A \backslash \Omega $.

16. By definition

$$ P(B | A) = \frac{P(A \cap B)}{P(A)} $$

and

$$ P(C | A \cap B) = \frac{P(A \cap B \cap C)}{P(A \cap B)} $$

Multiplying the above two equalities we get

$$ P(B | A) P(C | A \cap B) = \frac{P(A \cap B \cap C)}{P(A)} $$

Multiplying both sides by $P(A)$ we get

$$ P(A \cap B \cap C) = P(A) P(B | A) P(C | A \cap B) $$

This generalizes the multiplication rule to three events.

17. The multiplication rule can be generalized to an arbitrary number of events and then is known as the *Chain rule*

$$P(A_1 \cap A_2 \ldots A_n) = P(A_1) P(A_2 | A_1) P(A_3 | A_1 \cap A_2) \cdots P(A_n | \cap_{i=1}^{i=n-1} A_i)$$

18. Two events $A$ and $B$ are said to be *independent* if $P(A \cap B) = P(A) P(B)$. When $P(B) \neq 0$ this is equivalent to saying $P(A) = P(A | B)$ i.e. the occurrence of event $B$ does not affect the probability of event $A$. Events that are not independent are said to be *dependent*.

19. Two events $A$ and $B$ are said to be *conditionally independent* given $C$ when $P(A \cap B | C) = P(A | C) P(B | C)$.

20. *Bayes' Theorem* states

$$ P(B | A) = \frac{P(A \cap B)}{P(A)} = \frac{P(A | B) P(B)}{P(A)} $$

It is a trivial consequence of the definition of conditional probability but provides a valuable mathematical tool to update the probability $P(B)$ of an event $B$ (updated value being $P(B | A)$) given new evidence ($P(A)$) and the likelihood ($P(A | B)$) of event $A$ given $B$. Each of the terms in the theorem statement have a conventional name as below

$$ \mathrm{ Posterior } = \frac{\mathrm{ Likelihood } \cdot \mathrm{ Prior }}{\mathrm{ Evidence }} $$



21. The *Argmax* (arguments of the maxima) of a function $f$ on a set $D$ is the set of points in $D$ where $f$ attains its maximum value. Formally it is defined as

$$ \underset{x \in D}{\mathrm{arg\,max}} f(x) = \{ x | f(x) \geq f(y), \forall y \in D \} $$

Analogously *Argmin* (arguments of the minima) is defined as

$$ \underset{x \in D}{\mathrm{arg\,min}} f(x) = \{ x | f(y) \geq f(x), \forall y \in D \} $$

22. The right hand side denominator $P(A)$ in Bayes' Theorem is a *normalizing constant* that ensures that the probability condition $P(\Omega)=1$ is fulfilled. This denominator can be ignored when computing which events out of a set are more likely since it is the same for all cases. In particular

$$ \underset{B}{\mathrm{arg\,max}} \frac{P(A | B) P(B)}{P(A)} = \underset{B}{\mathrm{arg\,max}} P(A | B) P(B) $$

23. The *Law of Total Probability* states that if $B_n, i=1,2,\ldots$ is countable set of mutually exclusive ($B_i \cap B_j = \emptyset$ if $i \neq j$) and collectively exhaustive ($\cup B_i = \Omega$) events, then for any event $A$

$$ P(A) = \sum_n P(A \cap B_n) = \sum_n P(A | B_n) P(B_n) $$

24. The law of total probability may be used to write a more general form of *Bayes' Theorem*

$$ P(B_j | A) = \frac{P(A | B_j) P(B_j)}{P(A)} = \frac{P(A | B_j) P(B_j)}{\sum_{i=1}^n P(A | B_i) P(B_i)} $$

where $A \subseteq \cup_{i=1}^n B_i, P(A) > 0$, and $B_i \cap B_j = \emptyset$ for $i \neq j$.

25. A *Random Variable* is a function $X : \Omega \to \mathbb{R}^n$ (typically $n=1$). It is a mathematical formalization of an object or a quantity that depends on random (*stochastic*) events. The numerical ranges of random variables are more conducive to mathematical analysis rather than working with events themselves.

26. A *Discrete Random Variable* is function $X : \Omega \to S$ where $S$ is a countable subset of $\mathbb{R}^n$.

27. An *Indicator Random Variable* or a *Bernoulli trial* is a random variable $X : \Omega \to \{0, 1\}$ where the range is the set $\{0, 1\}$.

28. The *Probability Mass Function* (pmf) $p(x)$ of a discrete random variable $X$ assigns probabilities to values $x$ in the range of the random variable.

$$ p(x) = p(X = x) = P(A_x) $$

where $A_x = \{\omega \in \Omega : X(\omega) = x\}$

When a random variable $X$ is distributed according to pmf $p(x)$ this is denoted as $X \sim p(x)$. An alternative notation for the pmf associated with a random variable $X$ is $p_X(x)$. Any probability mass function must satisfy the following two conditions

- $$ \sum_i p(x_i) = \sum_i P(A_{x_i}) = P(\Omega) = 1 $$
- $$ p(x) \geq 0 $$

29. For any two random variables $X$ and $Y$ the *Joint Probability Distribution* $p(x, y)$ is defined to be

$$ p(x, y) = p(X = x \cap Y = y) $$

30. Given a joint probability distribution $p(x, y)$ the *Marginal Probability Distributions* are defined to be

$$ p_X(x) = p(X = x) = \sum_y p(x, y) $$

$$ p_Y(y) = p(Y = y) = \sum_x p(x, y) $$

If $X$ and $Y$ are independent then $p(x, y) = p_X(x)  p_Y(y)$ but in general the marginal distributions do not determine the joint distribution.

31. The *expectation* $E(X)$ or *mean* or *average* of a random variable $X \sim p(x)$ is given by

$$ E(X) = \sum_x x p(x) $$

which is well defined when $\sum_x |x| p(x) < 0$ where $|x|$ is the absolute value of $x$.

32. Let $Y \sim p(y)$ be a random variable and $g(Y)$ be any function. Then if $E(g(Y))$ is defined

$$ E(g(y)) = \sum_y g(y) p(y) $$

33. *Properties of Expectation*

- For all $X$ and $Y$

$$\begin{align*}
E(X + Y) =& \sum_{x,y} (x + y) p(x, y) \\
         =& \sum_x \sum_y x p(x, y) + \sum_x \sum_y y p(x, y) \\
         =& \sum_x x \sum_y p(x, y) + \sum_y y \sum_x p(x, y) \\
         =& \sum_x x p_X(x) + \sum_y y p_Y(y) \\
E(X + Y) =& E(X) + E(Y)
\end{align*}$$

- If $X$ and $Y$ are independent

$$\begin{align*}
E(X Y) =& \sum_{x,y} x y p(x, y) \\
       =& \sum_x \sum_y x y p_X(x) p_Y(y) \\
       =& \sum_x x p_X(x) \sum_y y p_Y(y) \\
E(X Y) =& E(X) E(Y)
\end{align*}$$

34. The expected value is invariant under linear transformation. For any $a$, $b$ and a random variable $X \sim p(x)$

$$\begin{align*}
E(a + bX) =& \sum_x (a+bx)p(x) \\
          =& \sum_x a p(x) + \sum_x b x p(x) \\
          =& a \sum_x p(x) + b \sum_x x p(x) \\
E(a + bX) =& a + b E(X)
\end{align*}$$

35. The *Variance* $Var(X)$ of a random variable $X$ is the average deviation from its expected value.

$$\begin{align*}
Var(X) =& E([X - E(X)]^2)\\
       =& E(X^2 - 2 X E(X) + E(X)^2) \\
       =& E(X^2) - 2 E(X) E(X) + E(X)^2 \\
Var(X) =& E(X^2) - E^2(X)
\end{align*}$$

36. For any $a$, $b$ and a random variable $X$

$$\begin{align*}
Var(a + bX) =& E([a + bX - E(a + bX)]^2) \\
            =& E([a + bX - a - bE(X)]^2) \\
            =& E(b^2 [X - E(X)]^2) \\
            =& b^2 E([X - E(X)]^2) \\
Var(a + bX) =& b^2 Var(X)
\end{align*}$$

37. For any random variable $X$

$$\begin{align*}
E(X-E(X)) =& E(X) - E(E(X)) \\
          =& E(X) - E(X) \\
E(X-E(X)) =& 0
\end{align*}$$

Now suppose $X$ and $Y$ are two independent random variables and let $U=X-E(X)$ and $V=Y-E(Y)$ be two additional random variables defined in terms of $X$ and $Y$. By definition $U$ and $V$ are independent if for any given values $u$ and $v$, $P(U=u \cap V=v) = P(U=u) \cdot P(V=v)$. Since the expectation of a random variable is a constant $P(U=u)=P(X-E(X)=u)=P(X=u+E(X))$ and $P(V=v)=P(Y-E(Y)=v)=P(Y=v+E(Y))$. Hence for any two independent random variables $X$ and $Y$, the random variables $U=X-E(X)$ and $V=Y-E(Y)$ are also independent.

Using the above two results the variance of a sum of two independent random variables may be decomposed into the sum of their variances as shown below.

$$\begin{align*}
Var(X + Y) =& E([X+Y - E(X+Y)]^2) \\
           =& E([X + Y - E(X) - E(Y)]^2) \\
           =& E([X-E(X) + Y-E(Y)]^2) \\
           =& E([X-E(X)]^2 + 2[X-E(X)][Y-E(Y)] + [Y-E(Y)]^2) \\
           =& E([X-E(X)]^2) + 2E([X-E(X)][Y-E(Y)]) + E([Y-E(Y)]^2) \\
           =& E([X-E(X)]^2) + 0 + E([Y-E(Y)]^2) \\
Var(X + Y) =& Var(X) + Var(Y)
\end{align*}$$

38. The *Covariance* $Cov(X, Y)$ of two random variables $X$ and $Y$ is the expected value of the product of deviations from their respective means.

$$ Cov(X, Y) = E([X-E(X)][Y-E(Y)]) $$

39. The *Standard Deviation* is defined to be the square root of the variance.

40. By convention for a given distribution the mean is denoted by $\mu$, the variance as $\sigma^2$ and the standard deviation as $\sigma$.

41. The *Conditional Probability Distribution* $p_{X|Y} (x|y)$ is defined as

$$ p_{X|Y} (x|y) = \frac{p(x, y)}{p_Y(y)} $$

for $y$ such that $p_Y(y) > 0$.

42. The *Chain Rule for Random Variables* can be written in terms of conditional probability distributions for example

$$ p(w, x, y, z) = p(w) p(x | w) p(y | w, x) p(z | w, x, y) $$

43. Empirically the proportion of times $C(u)$ a certain event $u$ occurs is called the relative frequency $f_u$ and is given by

$$ f_u = \frac{C(u)}{N} $$

where $N$ is the total number of trials.

44. A common problem is how to estimate the probability function $P$ in practice. There are two broad approaches to this problem

- *Parametric Methods* : This approach assumes the data generating process conforms to some particular parameterized family of probability distributions. The task of estimating the required probability function $P$ is then reduced to estimating the parameters of the assumed probability distribution that fit the data well. Since there are typically only a few parameters for most commonly used probability distribution the amount of data needed for a good fit tends to be low.

- *Non-parametric or distribution-free Methods* : Empirically a discrete probability distribution $P$ may be estimated by counting a large number of random events. From this a continuous distribution may be obtained through interpolation. Since such non-parametric methods provide little prior information they tend to require much more training data.

45. By convention when denoting a well know parameterized probability distribution the random variable arguments of the distribution precede the parameters and are separated by a semicolon (;). For example the Binomial distribution is denoted as $b(r; n p)$ where $r$ is the random variable, $n$ and $p$ are the parameters.

46. A trial (experiment) that can have only two outcomes (say 0 and 1) is known as a *Bernoulli trial*.

47. The probability distribution for the number of successes in a series of independent Bernoulli trials is know as the *Binomial distribution* $b(r; n, p)$. Specifically the binomial distribution $b$ computes the probability for $r$ successes out of $n$ trials where the probability of success in any trial is $p$.

$$ b(r; n, p) = {n \choose r} p^r (1-p)^{n-r} $$

where

$$ {n \choose r} = \frac{n!}{(n-r)! r!} $$

for $0 \leq r \leq n$.

48. Let $X$ be a random variable that only takes on values 1 or 0 with probability $p$ or $q=(1-p)$ respectively. Suppose $X_i$ are random variables representing the outcome of a series of independent, identically distributed Bernoulli trials indexed by $i$. The expected value $E(X_i)$ of single trial (say $i=1$) is by definition

$$ E(X_1) = (1)(p) + (0)(q) = p $$

By linearity of expected values

$$ E(\sum_i X_i) = \sum_i E(X_i) = np $$

Similarly the variance of a single trial is

$$\begin{align*}
Var(X_1) =& E(X_1^2) - E^2(X_1)\\
         =& p - p^2 \\
         =& p(1-p) \\
         =& pq
\end{align*}$$

By independence of $X_i$'s

$$ Var(\sum_i X_i) = \sum_i Var(X_i) = npq $$

In summary the binomial distribution $b(r; n, p)$ has an expectation $np$ and variance $np(1-p)$.

49. The *Normal or Gaussian distribution* $n(x; \mu, \sigma)$ is commonly used model the average value of many observations of a real valued random variable. The two parameters $\mu$ and $\sigma$ are the mean and standard deviation of the distribution.

$$ n(x; \mu, \sigma) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x-\mu)^2}{2\sigma^2}} $$

50. The Normal distribution may be used (with caution) to approximate a binomial distribution for a series of Bernoulli trials where both outcomes are roughly equal or the number of trials is very large i.e. when $np(1-p) > 5$.
